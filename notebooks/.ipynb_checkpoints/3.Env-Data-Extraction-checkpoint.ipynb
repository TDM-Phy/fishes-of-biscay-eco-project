## 3. Environmental data extraction and cleaning ##

This script will cover the addition of environmental variables to our fish dataset allowing us to model the influence of these variables on fish count and family abundance. These variables can be broken down into two categories static and dynamic, my static data will not require any specific date matching, the dynamic will due to its nature. For each variable I will outline below its relevance, method of extraction and the source of its data. 

### Static (no date) Code ###

- Bathymetry (depth) — GEBCO 15-arcsec global grid.  
Why: Depth is a core habitat driver; we’ll also derive slope.  
How: Download a clipped GeoTIFF to cover my study area.  
Source: GEBCO gridded bathymetry

- Distance to coast — Natural Earth land polygons.  
Why: Shelf–offshore gradient proxy.  
How: Calculate nearest distance (km) from points to coastline rasters.  
Source: Natural Earth 1:10m land.

### Dynamic (date-matched) ###

- Sea Surface Temperature (SST)
Why: Temperature is the dominant seasonal/latitudinal driver.  
How: Pull monthly SST via ERDDAP griddap (OPeNDAP) subset over my study area and time period  
Sources: OISST product page; ERDDAP griddap endpoints

- Chlorophyll-a (CHL)   
Why: Proxy for primary productivity, has significance for food web interactions  
How: Monthly ERDDAP griddap subset over my study area and time period
Sources: CoastWatch ERDDAP CHL datasets; Copernicus Marine chlorophyll product.

I'll Start with static....

GEBCO provides an online tool to allow you to download bathymetry in TIFs which you can specify you're study area. Takes 2 seconds and we have a relevant bathymetry tif to go "bathymetry'


### Step 1. Bathymetry ###

```python
import rioxarray as rxr
import xarray as xr
import pandas as pd
from pyproj import Transformer
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
from shapely.geometry import box
import numpy as np
import os
from pathlib import Path
from pandas.tseries.offsets import MonthEnd
import requests

# Specify the location of my bathymetry tif I downloaded

bathy_path = "data/data_raw/bathymetry.tif"

# Load it in using rioxarray to help visualise our tif

bathy = rxr.open_rasterio(bathy_path).squeeze()

# Quick plot to visualise

bathy.plot(figsize=(8,6), cmap="Grays_r")

# I'll load in my data

fish_df = pd.read_csv("data/data_interim/obis_fish_regions.csv")

# assign your coordinates to a points array

lon = fish_df["decimalLongitude"].to_numpy()
lat = fish_df["decimalLatitude"].to_numpy()

# Now ensure that both the raster I imported and my obis data are on the same CRS system. to do this ask what crs is raster in 

raster_crs = bathy.rio.crs

# Output will be the CRS the raster is in, then specify what the obis data is in which is

wgs86 = "ESPG:4326"

# Now I'll right an if function to ensure the raster and OBIS will be on the same CRS function. 
# the transformer function from pyproj will be required

if str(raster_crs).upper() != "EPSG:4326":
    transformer = Transformer.from_crs(wgs84, raster_crs, always_xy=True)
    x, y = transformer.transform(lon, lat)
else:
    x, y = lon, lat

# Now its all in the same CRS format, build a datsaet (points_df) of the points using xarray

points_df = xr.Dataset(coords={"x": ("points", x), "y": ("points", y)})

# Now I can cross reference the records location to obtain a depth from the raster

depth_value = bathy.sel(x=points_df["x"], y=points_df["y"], method="nearest").values

# I'll now make a column in my OBIS data to allocate the depth values

fish_df["Ocean_depth_m"] = depth_value

# Now have a quick look to see it went well

fish_df[["decimalLatitude", "decimalLongitude", "region", "Ocean_depth_m"]].head()

# In a perfect world we would now have a column full of negative figures. I'll run a QC check to be sure

print("Any values that suggest elevation over depth are here:", (fish_df["Ocean_depth_m"] >= 0).mean())

# Pandas contains a nice Describe function which can give you an instant idea of the data contained in a column
# I usually use it to make sure no wild value gets through when downloading data or computing a transformation

print("\nDepth summary (m):") 
print(fish_df["Ocean_depth_m"].describe())

# The describe definiton has brought back values suggesting elevation so I
# Count of positive (or zero) depths
num_positive = (fish_df["Ocean_depth_m"] >= 0).sum()
total_records = len(fish_df)

# I just want to check How many positive values are in the series as the describe function as found a few positive
print(f"Records with positive or zero depth: {num_positive:,} "
      f"({num_positive/total_records:.2%} of total)")

# 100 records have come up with positive values however, this can be due to the pixel size of the raster on coastal records
# I'll make a plot to determine if this is the case and if it is then I'll just drop them as its a negligible proportion. 

plt.figure(figsize=(6,6))
plt.scatter(fish_df["decimalLongitude"], fish_df["decimalLatitude"], 
            c=(fish_df["Ocean_depth_m"] >= 0), cmap="coolwarm", s=5)
plt.title("Positive-depth (land) points in red")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# Happy with the fact these are from the pixel size, so ill remove affected records from study as they are around 0.01%

fish_df = fish_df[fish_df["Ocean_depth_m"] < 0].copy()
print(f"Remaining records: {len(fish_df):,}")
```

After confirming that roughly 100 records had positive depth values, I inspected their coordinates and found they were located right along the coastline.
These are caveats from the bathymetry raster’s pixel resolution rather than true data points on land. Due to the negligible number of the records, I decided to remove all these points from study. 

### Step 2. Distance to coast ..... ###

```python
# This step can be donw externally via the natural earth website, you get a couple of shapefiles 

# Load the shapefile
land = gpd.read_file("data/data_raw/ne_10m_land.shp")

# Plot to confirm
land.plot(color="lightgray", edgecolor="black", figsize=(6,4))
plt.title("Natural Earth's Land shapefile")
plt.show()

# Excellent pretty map achieved. now to use it
# This next step crashed my server a couple of times so I'm adding in a step to clip my land shapefile to my study area 

lon_min, lon_max = fish_df["decimalLongitude"].min(), fish_df["decimalLongitude"].max()
lat_min, lat_max = fish_df["decimalLatitude"].min(), fish_df["decimalLatitude"].max()

# add a small margin so coastline just outside the box is included
margin_deg = 1.0
study_area_wgs84 = box(lon_min - margin_deg, lat_min - margin_deg,
                       lon_max + margin_deg, lat_max + margin_deg)
study_area_gdf = gpd.GeoDataFrame(geometry=[study_area_wgs84], crs="EPSG:4326")

# Now clip the shapefile to study area

land_cliped = gpd.clip(land, study_area_gdf)

# First I need to make a geodataframe of my fish_data records in WGS84

g_fish_df = gpd.GeoDataFrame(
    fish_df.copy(),
    geometry=gpd.points_from_xy(fish_df["decimalLongitude"], fish_df["decimalLatitude"]),
    crs="EPSG:4326"
)

# Being from the UK I prefer everything (accept car speed limits) to be in metric so I'll convert the crs to a metric one

g_fish_df_m = g_fish_df.to_crs(epsg=3035)
land_m = land_cliped.to_crs(epsg=3035)

# Now this is where we hit quite a few roadblocks in my prior Queensland fisheries research. 
# you need to specify that all land in youre study area is one entity otherwise 
# it has a funny habit of only using one polygon to determine as land, shouldnt be an issue with this project but just in case

Lands_of_study = land_m.union_all()

# Now we can use lambda to calculate the distance of each record to nearest land in km

g_fish_df["Distance_from_coast_km"] = g_fish_df_m.geometry.apply(lambda p: p.distance(Lands_of_study) / 1000.0)
fish_df["Coastal_distance_km"] = g_fish_df["Distance_from_coast_km"]

# Use describe again for a qc check
print(fish_df["Coastal_distance_km"].describe())
```

### Step 3. SST ###

I'll start with Sea Surface Temp. NOAA have a database ERDDAP which actually has API's like OBIS which streamlines the process. I'll go through it below. Before we start, we need to define our SST snapshot. Due to the oceans natural ability to regulate temperature and the study area being off shore, I'll be looking at SST in a monthly snapshot (pretty common for studies involving SST). 

```python
# Ensure my Date column in the fish data is in date format
fish_df["eventDate_dt"] = pd.to_datetime(
    fish_df["eventDate_dt"],
    errors="coerce",   
    dayfirst=False      
)

# Add a key to each record in order to add its monthly SST. Bascially make a column to say month , year 
fish_df["Month_Year"] = fish_df["eventDate_dt"].dt.to_period("M").dt.to_timestamp()

# Outline my study area 
LON_MIN, LON_MAX = -10.5, -1.0
LAT_MIN, LAT_MAX = 36.0, 48.5

# Now I'm going to call the website url/ path in order to access my dataset. Thanks to issues regarding both Trump and ERDDAP. 
# I'll be using met office which any UK resident will frown at but its got SST and I need SST
HADISST_PATH = Path("data/data_raw/HadISST_sst.nc")

# Now open the SST dataset 
SST_ds = xr.open_dataset(HADISST_PATH)

# I'm going to build a function that will help me search through this foreign dataset looking for specific columns by their contents
def first_existing(mapping, candidates):
    """Return the first key from candidates that exists in mapping."""
    lower = {k.lower(): k for k in mapping}
    for c in candidates:
        if c.lower() in lower:
            return lower[c.lower()]
    return None

# Search for my lat, long and time columns
lon_name = first_existing(SST_ds.coords, ["longitude", "lon"])
lat_name = first_existing(SST_ds.coords, ["latitude", "lat"])
time_name = first_existing(SST_ds.coords, ["time"])
if not (lon_name and lat_name and time_name):
    raise ValueError("Could not find lon/lat/time ")

# Define the Key varaible I want from this
sst_var = first_existing(SST_ds.data_vars, ["sst", "SST"])
if not (sst_var): 
    raise ValueError("Could not find SST")

# Now I'm planning to use a slice function, which can create issues as slicing assumes a ascedning or descending order. 
# as lats and longs in datasets dont usually follow such order it can cause the function to short circuit and return nothing
# Therefore a key step with this is to generate a function that bypasses this ordered slicing see below

def slice_anyorder(coord_values, a, b):
    """Return a slice that works whether coord is ascending or descending."""
    asc = bool(coord_values[0] < coord_values[-1])
    return slice(a, b) if asc else slice(b, a)


# Clip my SST dataset to study area
lat_slice = slice_anyorder(SST_ds[lat_name].values, LAT_MIN, LAT_MAX)
lon_slice = slice_anyorder(SST_ds[lon_name].values, LON_MIN, LON_MAX)
SST_study_area = SST_ds[[sst_var]].sel({lat_name: lat_slice, lon_name: lon_slice})

# Now I need to isolate all the months that I'll want SST for. 
months_for_SST = pd.Index(sorted(pd.to_datetime(fish_df["Month_Year"].values).unique()))

# Make vector holding the times in the SST dataset and develop an indexer 
hadisst_times = pd.DatetimeIndex(SST_study_area[time_name].values)
nearest_pos = hadisst_times.get_indexer(months_for_SST, method="nearest")


valid = nearest_pos >= 0
if (~valid).any():
    # If any months fall completely outside HadISST coverage, drop them
    print("Warning: Some months fall outside HadISST range:", months_for_SST[~valid].tolist())
    months_for_SST = months_for_SST[valid]
    nearest_pos = nearest_pos[valid]

# Pull only those monthly slices once; label them with 'stack_time' = your months
sst_stack = SST_study_area[sst_var].isel({time_name: xr.DataArray(nearest_pos, dims="stack_time")})
sst_stack = sst_stack.assign_coords(stack_time=("stack_time", months_for_SST))

# Normalize fish_data lat/lon/months to match dataset convention
fish_lon = fish_df["decimalLongitude"].to_numpy().astype(float)
fish_lat = fish_df["decimalLatitude"].to_numpy().astype(float)
fish_mo  = pd.to_datetime(fish_df["Month_Year"].to_numpy())

# Build indexers: one element per row (dimension 'points')
idx = {
    "stack_time": xr.DataArray(fish_mo, dims="points"),
    lat_name:     xr.DataArray(fish_lat, dims="points"),
    lon_name:     xr.DataArray(fish_lon, dims="points"),
}

# One batched lookup across all records using my indexers above (This is my joining step)
sst_samples = sst_stack.sel(**idx, method="nearest")  

units = str(SST_ds[sst_var].attrs.get("units", "")).lower()
sst_vals = sst_samples.values.astype("float64")
if "k" in units or np.nanmedian(sst_vals) > 100:
    sst_vals = sst_vals - 273.15

# Now add it to my fish data
fish_df["sst_c"] = sst_vals

# Run a few summary points to let me know how it went
print("\n=== SST (°C) summary ===")
print(fish_df["sst_c"].describe(percentiles=[0.05, 0.5, 0.95]))
print(f"Missing SST fraction: {fish_df['sst_c'].isna().mean():.2%}")

# Number of records left without SST ~0.16%, Im taking that as a win. remove any without sst before proceeding
fish_df = fish_df.dropna(subset=["sst_c"]).reset_index(drop=True)

# Save new dataset into my Data folder  
out_csv = Path("data/data_interim/fish_with_sst.csv")
out_csv.parent.mkdir(parents=True, exist_ok=True)
fish_df.to_csv(out_csv, index=False)
```

### Step 4. SST Anomalies ###

But before I go to Chl-A, I want to add another column in which I will identify how each month of the study compared to its running average SST for that month. This can be a very crucial but transient variable. I originally understood this method working with coral looking for bleaching events but its appliable to fish and can help understand populations dynamics. 

```python
# First step here is defining my baseline. which depends on good annual representation of fish records.
# We know from my spatial varaition section in the last script runs from 1993-2024
Baseline_start = "1993-01-01"
Baseline_end = "2023-12-31"

# Subset HadISST to baseline window within the study area defined SST dataset I made before
ds_monthly_avg = SST_study_area.sel({time_name: slice(Baseline_start, Baseline_end)})

# Now I've got to get each months average SST, but the trick involves doing that with location
sst_monthly_avg = ds_monthly_avg[sst_var].groupby(f"{time_name}.month").mean(skipna= True)

# Now I'll build a 12 month stack that can be sampled from using month as teh index
months_12 = np.arange(1, 13, dtype=int)
SSTM_stack = sst_monthly_avg.sel(month=xr.DataArray(months_12, dims="stack_time")) \
                     .assign_coords(stack_time=("stack_time", months_12))

# Build my indexers to cross reference with the fish dataset
fish_lon = fish_df["decimalLongitude"].to_numpy().astype(float)
fish_lat = fish_df["decimalLatitude"].to_numpy().astype(float)
fish_month_nums = pd.to_datetime(fish_df["Month_Year"].to_numpy()).month

idx = {
    "stack_time": xr.DataArray(fish_month_nums, dims="points"),  # month number
    lat_name:     xr.DataArray(fish_lat,       dims="points"),
    lon_name:     xr.DataArray(fish_lon,       dims="points"),
}

# Using my described Indexers search the monthly SST stack I've built
sst_baselines = SSTM_stack.sel(**idx, method="nearest").values.astype("float64")


# Now I have my baselines. attached them to fish df then workout the difference between actuals
fish_df["sst_baseline_c"] = sst_baselines
fish_df["sst_anom_c"] = fish_df["sst_c"] - fish_df["sst_baseline_c"]

# Run a few summarys to understand how it came out
print("\n=== SST anomaly (°C) summary ===")
print(fish_df["sst_anom_c"].describe(percentiles=[0.05, 0.5, 0.95]))
print(f"Missing climatology fraction: {fish_df['sst_baseline_c'].isna().mean():.2%}")
```

### Step 5. Chlorophyll ###

This process will be similar to SST pipeline. Essential I'll be downloading a dataset from some third party. Be presented with a gridded monthly Chl-A dataset. Clip it to my study area and time period. Finally attaching it to my dataset. 
Only took me 2-3 night to figure out SST so lets see if this will be any faster. 

```python
# With SST unforntunately I had to settle for manual donwload of the data set. 
# Not the end of the world but data scientists/ engineers are required to use APIs to construct automated workflows
# So if any employers did notice that before, please witness me using APIs for everything else. 

url = "https://data.cci.ceda.ac.uk/thredds/dodsC/esacci.OC.mon.L3S.CHLOR_A.multi-sensor.multi-platform.MERGED.6-0.geographic"

# Now construct the chl dataset. 
Chl_ds = xr.open_dataset(url)

# Identify Lat / Lon / time
time_name = "time"
lon_name = "lon" if "lon" in Chl_ds.coords else "longitude"
lat_name = "lat" if "lat" in Chl_ds.coords else "latitude"

# This came about from trial and error: Build a function that can convert 0 - 360 degrees to -180 - 180 degrees
def wrap_lon_to_180(lon_array):
    return ((np.asarray(lon_array) + 180.0) % 360.0) - 180.0

# Now I'll pull over longitudes first and use the above function to ensure they are in correct format
Chl_ds = Chl_ds.assign_coords({lon_name: wrap_lon_to_180(Chl_ds[lon_name])}).sortby(lon_name)

# I'll define my study area. Done multiple times in previous cells I know but run this cell clean
LON_MIN, LON_MAX = -10.5, -1.0
LAT_MIN, LAT_MAX = 36.0, 48.5

# I'll now use slice to in order to reduce the chl dataset size to just relevant time and locations
# As explained in my SST pipeline this will require a function to slice without order

def slice_anyorder(coord_values, a, b):
    """Return a slice that works whether coord is ascending or descending."""
    asc = bool(coord_values[0] < coord_values[-1])
    return slice(a, b) if asc else slice(b, a)

lat_slice = slice_anyorder(Chl_ds[lat_name].values, LAT_MIN, LAT_MAX)
lon_slice = slice_anyorder(Chl_ds[lon_name].values, LON_MIN, LON_MAX)

Chl_Survey_area = Chl_ds.sel({lat_name: lat_slice, lon_name: lon_slice})

############################## Next stage is to identify the key variable for chlorophyll in the dataset. #####################

#  Run an if loop to search column titles for chlor a (most commonly used term for chl in scientific monitoring
chl_var = "chlor_a" if "chlor_a" in Chl_Survey_area.data_vars else list(Chl_Survey_area.data_vars)[0]
print(f"Chl variable: {chl_var}")
print("Units:", Chl_Survey_area[chl_var].attrs.get("units", ""))

# Confirm time coverage
chl_dates = Chl_Survey_area[time_name].to_index()
print("CHL time range:", chl_dates[0].date(), "→", chl_dates[-1].date())

### I now need to match the two datasets by time, this is done by building a date index from my fish dataset
Months_of_study = (
    pd.to_datetime(fish_df["Month_Year"])
    .dt.to_period("M").dt.to_timestamp()
    .sort_values().unique()
)
Months_of_study = pd.DatetimeIndex(Months_of_study)
Chl_dates = Chl_Survey_area[time_name].to_index()

# I create a vector called nearest_position which basiscally maps each fish data month to the closest date in chl a data
Nearest_position = Chl_dates.get_indexer(Months_of_study, method="nearest")

####################### I now need to build my stack of relevant Dates (months) from the chlor dataset. #######################

# Quite a few processes have been run in this code block so for trouble shooting I'll add in a few prints to know where I am at
print("Building susbet from OPenDap")

chl_dates_stack = (
    Chl_Survey_area[chl_var]
    .isel({time_name: xr.DataArray(Nearest_position, dims="stack_time")})
    .assign_coords(stack_time=("stack_time", Months_of_study))
    .transpose("stack_time", lat_name, lon_name)  # nice predictable order
).load()

############################## Date and coordinates are aligned and the new dataset has been clipped to survey area ############ 

# I now need to build my 3 coordinates that will search through the chl dataset and find the nearest coordinate with chl a value
fish_lon = wrap_lon_to_180(fish_df["decimalLongitude"].to_numpy().astype(float))
fish_lat = fish_df["decimalLatitude"].to_numpy().astype(float)
fish_month  = pd.to_datetime(fish_df["Month_Year"].to_numpy()).astype("datetime64[ns]")

idx = {
    "stack_time": xr.DataArray(fish_month, dims="points"),
    lat_name:     xr.DataArray(fish_lat, dims="points"),
    lon_name:     xr.DataArray(fish_lon, dims="points"),
}

# Use my coordinates to generate values for chl for each record
chl_per_point = chl_dates_stack.sel(**idx, method="nearest")
# Generate a numpy array the same length as my fish data
chl_value_pp = chl_per_point.values.astype("float64")

# Remove any negatives
chl_value_pp[chl_value_pp < 0] = np.nan

# Add to fish df
fish_df["chl_mg_m3"] = chl_value_pp

print("\n=== Chlorophyll (mg m^-3) summary ===")
print(fish_df["chl_mg_m3"].describe(percentiles=[0.05, 0.5, 0.95]))
print(f"Missing chlorophyll fraction: {fish_df['chl_mg_m3'].isna().mean():.2%}")

# Save complete dataset to csv in data folder
out_csv = Path("data/data_interim/fish_with_sst_chl.csv")
out_csv.parent.mkdir(parents=True, exist_ok=True)
fish_df.to_csv(out_csv, index=False)
```

I've now got a dataset of fish records and have enriched it with Sea Surface Temperatures, presence on thermal anomalies and now the chlorophyll for the month and location of each record (a proxy for primary productivity. Important things to note is that the chlorophyll data only runs from 09/1997 to 12/2022.
