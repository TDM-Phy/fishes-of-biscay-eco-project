## 4. Data audit and investigation ##

Alot of what has occured so far in the project is down to data engineers to create strong, idempotent datasets from a variety of data sources. The work of Data scientists usually starts now, however I dont have any data engineers to hand so I've completed the first stage myself. 
I now have a dataset containing discreet data in the form of fish records (presence), the family of the fish observed, the coordinates and date of observation. This has been matched to continuous data in the form of bathymetry, distance to shore, Sea Surface Temperature (SST) and Chlorophyll. 
I'll now build a pipeline that will run the checks on data as if it has just arrived. I will aim to :
- Validate the datas integrity
- Ensure spatial and temporal data is in uniform format
- Search for missing data and outliers
- Define analytical varaibles (Region, month and family)
- Understand the independant variable
- Check for collinearity
- Produce a clean master set ready for modelling


```python
import numpy as np
import pandas as pd
from datetime import datetime
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
import patsy 

################################# Clean up the master data sheet/ create my focus dataset ####################################

master_df = pd.read_csv("data/data_interim/fish_with_sst_chl.csv")

# Define study area and period 
study_area = {
          "lon_min" : -10.5, "lon_max" : -1.0, 
          "lat_min" : 36.0, "lat_max" : 48.5,
}

Period_start = pd.Timestamp("1997-09-01")
Period_end = pd.Timestamp("2022-12-31")

# Search for expected columns and assign simple identifier. 
# I have a bad habit of adding uppercase letters into titles, good to remove early on so assign cols for shorthand
cols = {
    "sp.name" : "scientificName",
    "family": "family", 
    "lat" : "decimalLatitude",
    "lon" : "decimalLongitude",
    "month_year" : "Month_Year",
    "eventdate" : "eventDate",
    "fish_depth" : "depth",
    "source" : "datasetName",
    "region": "region",
    "coast_km" : "Coastal_distance_km",
    "ocean_depth" : "Ocean_depth_m",
    "sst_actual" : "sst_c",
    "sst_month_bs" : "sst_basline_c",
    "sst_anomaly" : "sst_anom_c",
    "chl": "chl_mg_m3",
    "ind_count": "individualCount",
}

# Now I've selected and renamed my rows I'll subset to remove all unecessary columns
target_cols = [v for v in cols.values() if v in master_df.columns]
focus_df = master_df[target_cols].copy()

# Now rename my columns using the shorthand
rename_map = {v: k for k, v in cols.items() if v in master_df.columns}
focus_df = focus_df.rename(columns=rename_map)

#focus_df.head(10)

# Somewhere around now a data scientist would go through the steps ensuring lat longs date units are all in standardised format
# Fortunately I've already done all that so I wont redo , I'll just move on

## I will audit my dataset to understand if certain variables are missing alot of data
# Build a function that will handle number of na's and data coverage
def audit_nas_and_view_coverage(df):
    na_by_col = df.isna().mean().sort_values(ascending=False).rename("nas_pct")

    by_region = df.groupby("region").size().rename("n_records").sort_values(ascending=False)
    by_month  = df.groupby("month_year").size().rename("n_records")
    by_reg_mon = df.groupby(["region","month_year"]).size().rename("n_records")

    # Now I'll add in some descriptive stats for my continuouos variables
    conts = [c for c in ["sst_actual","sst_month_bs","sst_anomaly","chl",
                         "fish_depth","ocean_depth","coast_km"] if c in df.columns]
    desc = df[conts].describe(percentiles=[0.01,0.25,0.5,0.75,0.99]).T

    return na_by_col, by_region, by_month, by_reg_mon, desc

na_by_col, by_region, by_month, by_reg_mon, cont_desc = audit_nas_and_view_coverage(focus_df)

print("NAs by column")
print(na_by_col.head(8))

print("Records per region")
print(by_region.head(10))

print("Time span & monthly coverage")
print(by_month.index.min(), "→", by_month.index.max(), "| months:", by_month.size)

print("Continuous variables (percentiles)")
print(cont_desc[["mean","std","1%","25%","50%","75%","99%"]])
```

## Understanding the independant variable. ##

I can imagine the next section might confuse readers not familiar with the marine space, so I'll quickly explain as its part of my background studies, although I imagine the concept is very transferable across different data industries. 
Basically, if I leave abundance as it is, we very well could be just viewing greater abundance in May in the Bay of Biscay and assuming that is due to ecological reasons e.g. more food in this location in this month. However, later on someone reviewing the method found that in the year 80% of all surveys were conducted in August, and 70% of all studies were in the bay of biscay. Meaning you are observing differences in sampling effort not abundance. A crucial step therefore is calibrating your independant variable for unit effort. 
This is for obvious reasons not so straight forward with this dataset. Many different databases and no 'surveyID' key. So I'm going to build a surrogate version of sample ID. This will not replicate true effort but it will suffice for the purpose of this study. I'm going to define 1 survey as Lat,Lon (rounded to 3 figures to make a coordinate) x eventDate(still at dd-mm-yyyy accuracy) x datasetName. This will allow me to understand how many records were observed by each dataset in the same atime and location as 1 survey. 

```python
# Make a function that will build a surveyID key to use
def build_unique_surveys (
    focus_df,
    lat_col="lat",
    lon_col="lon",
    dataset_col="source",
    date_col="eventdate", 
    lat_round=3,
    lon_round=3
):
    """
    Create a surrogate survey identifier
    """
    df = focus_df.copy()
    day = pd.to_datetime(df[date_col], errors="coerce").dt.floor("D") 
    df["date_day"] = day
    df["lat_r"] = pd.to_numeric(df[lat_col], errors="coerce").round(lat_round)
    df["lon_r"] = pd.to_numeric(df[lon_col], errors="coerce").round(lon_round)
    surveyid = [dataset_col, "date_day", "lat_r", "lon_r"]
    df["surveyID"] = df[surveyid].astype(str).agg("|".join, axis=1)

    return df

focus_df = build_unique_surveys(focus_df)
```

### Collinearity ###

```python
# Ensure our region column is complete
regions = sorted(focus_df["region"].dropna().unique().tolist())
print("Regions found:", regions)

# Understanding each of my regions before I model within them
def region_qc(regions_df):
    by_month = regions_df.groupby("eventdate").size()
    by_source = regions_df.groupby("source").size() if "source" in regions_df.columns else pd.Series(dtype=int)
    print("  n records:", len(regions_df), "| months:", by_month.size)
    if not by_source.empty:
        print("  top sources:\n", by_source.sort_values(ascending=False).head(5))
    empty_recs = regions_df[["sst_actual","chl","coast_km","ocean_depth","ind_count"]].isna().mean().sort_values(ascending=False)
    print(" Empty records", empty_recs)

for r in regions:
    print(f"\n=== QC: {r} ===")
    region_qc(focus_df[focus_df["region"]==r])

# Now I'm treating my regions sepearately as described above. I'll now run through the checks before i can fit any models
## 1. Collinearity

def check_collinearity(
    focus_df, 
    region_name, 
    predictors=("sst_actual", "chl", "coast_km", "ocean_depth"),
):
    """ Zuur protocols, check for collinearity between predictors. 
    Due to chlor natural variation I'll log transform 
    Create Z scores for my predictor varaibles
    """
    df_by_region = focus_df.loc[focus_df["region"] == region_name].copy()
    preds = list(predictors)
    if "chl" in df_by_region.columns and "chl" in preds:
        df_by_region["log_chl"] = np.log(df_by_region["chl"].astype(float) + 1e-6)
        preds = [("log_chl" if p == "chl" else p) for p in preds]
    used = [p for p in preds if p in df_by_region.columns]
    Z = pd.DataFrame(index=df_by_region.index)
    for p in used:
        m = df_by_region[p].mean(skipna=True)
        s = df_by_region[p].std(skipna=True, ddof=0)
        Z[p + "_z"] = (df_by_region[p] - m) / s if (s and np.isfinite(s) and s > 0) else np.nan
        
    X = Z.dropna().astype(float)
    if X.empty:
        raise ValueError("No complete rows after z-scoring error.")
    corr_df = X.corr().round(2)
    Xv = sm.add_constant(X, has_constant='add')
    vif_df = pd.DataFrame({
        "variable": Xv.columns[1:],  # skip intercept
        "VIF": [variance_inflation_factor(Xv.values, i + 1) for i in range(len(X.columns))]
    }).sort_values("VIF", ascending=False).reset_index(drop=True)

    return corr_df, vif_df, used

# Now Print out a summary for each region to check collinearity 
region_name = "fr_atlantic" 
corr, vif, used_vars = check_collinearity(focus_df, region_name)
print("Predictors used:", used_vars)
print("\nPearson correlations:\n", corr)
print("\nVIF:\n", vif if vif is not None else "VIF unavailable (statsmodels not found)")

# Save a dataset here that I will use for my machine learning 
cleaned_path = "data/data_interim/cleaned_focus_dataset.csv"
focus_df.to_csv(cleaned_path, index=False)
print(f"Cleaned dataset saved to: {cleaned_path} (rows: {len(focus_df)})")
```

### Response distribution and overdispersion ###

In this step I really understand what GLM family will suit my response. I do that by defining the data type. Our data is shearly presence (no absence) we only have a record for presence. Without absence we are abit in limbo for attempting to model predictor varaibles against response. However, if we view each record as an event and not a record we can then aggregate events into time/space cells. We then understand for that cell what the number of records were, what the sst and chla was. This will not give us an abundance response but an event intensity and I can better understand the effort by determining the number of times that location cell was visited in a month. 
So now I'll build my space/time cells with event_instensity  and number of survey visits below

```python
# define the size of my space cells (I landed with 0.05 it gives cells about 21km^2 size)
grid_size = 0.05

# create a function that can create coordinate bins to make my grid
def assign_cells(d, lat='lat', lon='lon', res_deg=0.05):
    d = d.copy()
    d['lat_bin'] = np.floor(d[lat] / res_deg) * res_deg
    d['lon_bin'] = np.floor(d[lon] / res_deg) * res_deg
    d['cell_id'] = d['lat_bin'].round(4).astype(str) + "_" + d['lon_bin'].round(4).astype(str)
    return d

# assign cells with the chosen resolution

gridded_df = assign_cells(focus_df,res_deg=grid_size)

# I now define my response and effort variables in the new gridded df
event_counts = (gridded_df
            .groupby(['cell_id', 'month_year'], as_index=False)
            .size()
            .rename(columns={'size': 'event_counts'}))

effort = (gridded_df[['cell_id', 'month_year', 'surveyID']]
          .drop_duplicates()
          .groupby(['cell_id', 'month_year'], as_index=False)
          .agg(visits=('surveyID', 'nunique')))



cell_month = (event_counts
              .merge(effort, on=['cell_id', 'month_year'], how='outer')
              .fillna({'event_counts': 0, 'visits': 0})
              .astype({'event_counts': int, 'visits': int}))


# Next I'll attach my covariates to cells, using median for each 
covariates = ['sst_actual', 'chl', 'coast_km', 'ocean_depth']
predictor_values = (gridded_df[['cell_id', 'month_year'] + covariates]
           .groupby(['cell_id', 'month_year'], as_index=False)
           .median())

# Now attach the two tables to get one table with all necessary data for modelling

gridded_cell_df = (cell_month
       .merge(predictor_values, on=['cell_id', 'month_year'], how='left')
       .dropna(subset=covariates)
       .copy())

# Offset for effort
gridded_cell_df['offset_log_effort']= np.log(gridded_cell_df['visits'].astype(float))

## Now look at the distribution of the new response
response_summary = pd.DataFrame({
    'Loc cells'      : [len(gridded_cell_df)],
    'mean_event_intensity'      : [gridded_cell_df['event_counts'].mean()],
    'variation'       : [gridded_cell_df['event_counts'].var(ddof=1)],
    'var_to_mean' : [gridded_cell_df['event_counts'].var(ddof=1) / max(gridded_cell_df['event_counts'].mean(), 1e-12)],
    'zero_cells'   : [(gridded_cell_df['event_counts'] == 0).mean()],
})
print("Overall response summary:")
print(response_summary.to_string(index=False))

# I forgot to add my regins so I'll just add that for each cell using mode
cell_region_map = (
    gridded_df[['cell_id','region']].dropna()
    .groupby('cell_id')['region']
    .agg(lambda s: s.mode().iat[0])   
)
gridded_cell_df['region'] = gridded_cell_df['cell_id'].map(cell_region_map)

## I'm now going to fit a null poisson model (basically to test if the poisson data family is a good fit
observed_count = gridded_cell_df['event_counts'].astype(float).values
effort = np.exp(gridded_cell_df['offset_log_effort'].values)

# Compute null possion equation, im not going to write it out just look online
sum_observed, sum_effort = observed_count.sum(), effort.sum()
alpha_hat = np.log(max(sum_observed, 1e-12) / max(sum_effort, 1e-12))
poissons = np.exp(alpha_hat + gridded_cell_df['offset_log_effort'].values)

# pearsons dispersion
pearson = (observed_count - poissons) / np.sqrt(np.clip(poissons, 1e-12, None))
chi2 = float((pearson**2).sum())
df_resid = max(len(observed_count) - 1, 1) 
dispersion = chi2 / df_resid

# now look at how zero value fit in observed records and against model
obs_zero_prop = float((observed_count == 0).mean())
exp_zero_prop = float(np.mean(np.exp(-poissons)))


print("Overdispersion & Zeros for my poisson model")
print(f"Dispersion (Pearson chi2 / df): {dispersion:.2f}")
print(f"Observed zero proportion      : {obs_zero_prop:.3f}")
print(f"Expected Poisson zero prop    : {exp_zero_prop:.3f}")

### Very High dispersion, poisson is a poor fit
# Next up neg binomial

offset = gridded_cell_df['offset_log_effort'].astype(float).values
nb_count = gridded_cell_df['event_counts'].astype(float).values

# Apply a minimal Negbin model using statsmodel.apis, set up an intercept X

X = np.ones((len(nb_count), 1))

nb_model = sm.NegativeBinomial(nb_count, X, offset=offset).fit()
model_pred = nb_model.predict()

raw = float(nb_model.params[-1])
alpha_hat = raw if raw > 0 else float(np.exp(raw))

# NB variance
nb_var = model_pred + alpha_hat * (model_pred**2)

# Pearson residuals using NB variance
pearson_nb = (nb_count - model_pred) / np.sqrt(np.clip(nb_var, 1e-12, None))
dispersion_nb = float((pearson_nb**2).sum() / nb_model.df_resid)

print(f"Correct NB dispersion (Pearson χ²/df using NB var): {dispersion_nb:.2f}")
print(f"Estimated alpha (NB dispersion parameter): {alpha_hat:.3f}")
print(f"Baseline mean events per visit: {float(np.exp(nb_model.params[0])):.3f}")
```

### Baseline Negative Binomial Model

At this stage I ran a baseline Negative Binomial model following the Zuur et al. protocol.  
The goal here wasn’t to explain fish abundance yet, but simply to confirm that the distributional
assumptions of the model (variance structure) are appropriate before adding any predictors.

Up to this point, the Poisson model showed clear overdispersion:
- Variance-to-mean ratio ≈ 14  
- Pearson dispersion ≈ 14  
→ The data were far too variable for a Poisson, which assumes Var = Mean.

The Negative Binomial introduces a dispersion parameter (α) that allows the variance to scale, making it suitable for ecological count data where individuals are patchy or aggregated.

The results encapsulated in the output above confirms that the Negative Binomial family is appropriate for modelling fish detections in this dataset.
From here, I can safely move on to introduce minimal structure (region and/or month effects)
to capture broad spatial or temporal patterns before adding environmental predictors.

### Introducing Minimal Structure (Region & Month Effects)

The next step follows Zuur et al.’s recommendation to account for broad spatial or temporal trends *before* adding any continuous predictors.
The idea here is to see whether large-scale differences in fish detections across **regions** or **months** explain a meaningful portion of the remaining variance.  
If dispersion remains low after including these factors, it means the model structure is capturing the main grouping effects and the data are well-behaved for adding environmental variables later.
I also want to run a test to determine if my earlier suggestion to split up the dataset into the three regions is also suggested by running it in a model

```python
region_testing = gridded_cell_df.copy()

# Design matrix with region, patsy will create dummy varibales to turn my categorical region into a matrix
y, X = patsy.dmatrices('event_counts ~ C(region)', data=region_testing, return_type='dataframe')
counts = y.values.ravel()
offset = region_testing['offset_log_effort'].astype(float).values

# Fit NB (discrete) which estimates alpha
nb_reg = sm.NegativeBinomial(y, X.values, offset=offset).fit()
print(nb_reg.summary())

# Extract alpha 
raw = float(nb_reg.params[-1])
alpha_hat = raw if raw > 0 else float(np.exp(raw))

# Fitted means and NB-based dispersion overall
r_model_pred = nb_reg.predict()
var = r_model_pred + alpha_hat * (r_model_pred**2)
pearson = (counts - r_model_pred) / np.sqrt(np.clip(var, 1e-12, None))

# this was one of the trial and error moments that even after solving you dont fully understand
pearson_series = pd.Series(pearson, index=X.index)

disp_overall =  float((pearson_series**2).sum() / nb_reg.df_resid)
print(f"\nCorrect NB dispersion (overall): {disp_overall:.2f}   |   alpha ≈ {alpha_hat:.3f}")

# Per-region dispersion
per_region = []
for r, idx in region_testing.groupby('region').groups.items():
    pr = pearson_series.loc[idx].values              
    df_r = max(len(pr) - X.shape[1], 1)
    per_region.append((r, float((pr**2).sum() / df_r), len(pr)))


per_region = pd.DataFrame(per_region, columns=['region','nb_dispersion','n'])
print("\nPer-region NB dispersion:")
print(per_region.sort_values('nb_dispersion').to_string(index=False))
```

The output suggests that the 'West Iberia' and 'South Biscay' regions are both underdispersed in our model which makes sense when you view the two sample sizes involved. 'French Atlantic' region is still showing as slightly overdispersed but well within acceptable range so I can continue with the investigation using negative bionomial. The output also suggests that the regions can be kept pooled within the model and just add a coeffecient that handles my region category. 

### Adding in my continuous variables

Continuing with a negative bionomial model, we can now begin to test which environmental variables help to explain fish abundance. 

```python
# Just remembered that I wanted to log trasnform my chlorine, I ddi it before within a function but havent yet applied it 
gridded_cell_df['log_chl'] = np.log1p(np.clip(gridded_cell_df['chl'].astype(float), a_min=0, a_max=None))

# I'm going to standardise the continuous varaibles as the differences in magnitudes of the scales was causing issues
# Im going to try out a Standard scaler as its been suggested online 
from sklearn.preprocessing import StandardScaler

cont_vars = ['sst_actual', 'log_chl', 'coast_km', 'ocean_depth']

scaler = StandardScaler()
gridded_cell_df[[f'scd_{v}' for v in cont_vars]] = scaler.fit_transform(gridded_cell_df[cont_vars])

# I now write out my model formula
Neg_bin_model_formula = 'event_counts ~ scd_sst_actual + scd_log_chl + scd_coast_km + scd_ocean_depth + C(region)'

counts_df, X = patsy.dmatrices(Neg_bin_model_formula, data=gridded_cell_df, return_type='dataframe')
counts = counts_df.values.ravel()
offset = gridded_cell_df['offset_log_effort'].astype(float).values

# fit negative binomial with log-link
nb_model = sm.NegativeBinomial(counts, X.values, offset=offset).fit()
print(nb_model.summary())

## Need to correct alpha as its not setting automatically
raw = float(nb_model.params[-1])            
alpha_hat = raw if raw > 0 else float(np.exp(raw))


model_pred = nb_model.predict()
var = model_pred + alpha_hat * (model_pred**2)
pearson_nb = (counts - model_pred) / np.sqrt(np.clip(var, 1e-12, None))
dispersion_nb = float((pearson_nb**2).sum() / nb_model.df_resid)
print(f"\nCorrect NB dispersion (with covariates): {dispersion_nb:.2f}  |  alpha ≈ {alpha_hat:.3f}")
```

The initial Negative Binomial model was fitted across all regions (French Atlantic, South Bay of Biscay, and West Iberia) to assess general relationships between fish counts and environmental variables. While the model ran successfully , the results confirmed that regional effects dominated the variation in abundance.

This outcome was expected. The French Atlantic has much denser and more consistent survey coverage than the southern regions, meaning that the region factor largely captures differences in sampling design and effort rather than ecological gradients. In this pooled model, the between-region differences in observation density outweigh the finer-scale environmental effects that occur within each region.
To better understand the ecological drivers of abundance, it is more appropriate to model each region separately. 

In the next step, I’ll fit separate Negative Binomial models for each region to identify which environmental variables explain within-region variability in event counts and how their influence differs spatially.

```python
# In order to seperately model regions in quick succession I'll build a function to handle
# Unfortunately, I'm having a hard time reliably applying the dispersion coeefecient alpha so I''ll build a function to reliably locate and then use

# Create container for model results
region_models = {}
region_disp   = {}
region_alpha  = {}
region_results = []

# Loop through each region and fit a Negative Binomial model
for region, sub in gridded_cell_df.groupby('region'):
    print(f"\n=== {region.upper()} ===")

    # Define response, predictors, and offset
    counts = sub['event_counts'].values
    offset = sub['offset_log_effort'].values
    X = sub[['scd_sst_actual', 'scd_log_chl', 'scd_coast_km', 'scd_ocean_depth']]
    X = sm.add_constant(X)

    # Fit model
    nb_model = sm.NegativeBinomial(counts, X, offset=offset, loglike_method='nb2').fit()
    alpha = float(getattr(nb_model, "alpha", np.exp(nb_model.params[-1])))

    # Calculate dispersion 
    model_pred = nb_model.predict()  
    model_var = model_pred + alpha * (model_pred**2)
    resid = (counts - model_pred) / np.sqrt(np.clip(model_var, 1e-12, None)) 
    dispersion = float((resid**2).sum() / nb_model.df_resid)
    
    # Save fitted model + metrics
    region_models[region] = nb_model
    region_disp[region]   = dispersion
    region_alpha[region]  = alpha

    # Store model info
    region_results.append({
        'region': region,
        'n': len(sub),
        'dispersion': round(dispersion, 2),
        'alpha': round(float(nb_model.scale), 3)  # approximate NB α
    })

    # Print summary for review
    print(nb_model.summary().tables[1])
    print(f"\nDispersion: {dispersion:.2f} | alpha: {alpha:.3f}")

# Combine results into a summary table
region_summary = pd.DataFrame(region_results)
print("\n=== Per-region model summary ===")
display(region_summary)
```

The final step of this script will just be to clean up the above coeffecient tables. This will simulate making them ready to publish 

```python
# I'm going to build a P value significance heatmap and include some figures to observe the varaibles influence on the mean

# Call my predictors first
predictors = ['scd_sst_actual', 'scd_log_chl', 'scd_coast_km', 'scd_ocean_depth']

# Call my bin of model results i made before
records = []

# Loop through each regions model and extract the data needed for heatmap
for region, model in region_models.items():
    for var in predictors:
        beta = model.params.get(var, np.nan)
        pval = model.pvalues.get(var, np.nan)
        expb = np.exp(beta)
        records.append({
            'region': region,
            'variable': var,
            'beta': beta,
            'pval': pval,
            'exp_beta': expb,
            'perc_change': (expb - 1) * 100  # interpretive effect (% change)
        })

coef_summary = pd.DataFrame(records)
display(coef_summary)

# Now I have a tided coeffecient table, make a pivot to heatmap (Region x Variable)

chessboard = coef_summary.pivot(index='variable', columns='region', values='pval')
display(chessboard)

# Now I'll add a few visual parameters to better present significance 
import seaborn as sns
import matplotlib.pyplot as plt

binary_sig_fig = chessboard < 0.05
binary_sig_fig = binary_sig_fig.astype(int)

# Call specific colours to accentuate the binary result
cmap = sns.color_palette(["#d3d3d3", "#2ca02c"])

plt.figure(figsize=(6, 3))
sns.heatmap(binary_sig_fig, annot=chessboard.round(3), cmap=cmap, cbar = False,
            linewidths=0.5, fmt=".2f")
plt.title("Predictor Significance Across Regions")
plt.ylabel("Predictor")
plt.xlabel("Region")
plt.show()

# Finally I want to include some figures to help identify how each significant varaible relationship influences the mean
significant = coef_summary[coef_summary['pval'] < 0.05].copy()
significant['direction'] = np.where(significant['beta'] > 0, 'increase', 'decrease')

for _, row in significant.iterrows():
    print(f"In {row['region']}: +1 SD {row['variable']} → "
          f"{row['perc_change']:+.1f}% {row['direction']} in fish counts (p={row['pval']:.3f})")
```

### Model Limitations and Next Steps

The Negative Binomial models developed here provide a solid first assessment of which environmental predictors significantly influence fish event counts across the 3 regions of the study. However, I have merely knocked on the door of the statistical investigation of this dataset. The above model does not

- **Include interaction terms**, meaning potential conditional relationships between variables (e.g. SST × chlorophyll or depth × distance to coast) were not tested.  
- **Account for spatial nesting or temporal structure**, treating each grid cell and time step as independent observations. A more complete approach would include spatial random effects (e.g. by nesting latitude–longitude cells) and temporal covariates or random effects to control for spatial and temporal autocorrelation.  
- **Undergo model selection or goodness-of-fit comparison** using information criteria such as AIC or AICc (e.g. via an exhaustive subset approach like `dredge()` in R’s *MuMIn* package). These procedures help identify the most parsimonious model that balances fit and complexity.

Implementing these refinements would improve model realism and inferential accuracy, but each adds analytical depth and time.
