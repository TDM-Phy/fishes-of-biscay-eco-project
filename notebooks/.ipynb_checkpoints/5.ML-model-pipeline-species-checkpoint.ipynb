### 5b. ML data pipeline Species

Originally I completed this next step using a family level investigation. However the random forest accuracy came out low with diffculty predicting certain clusters with alot of overlap observed. After some thinking I decided that as all of my data was coming from the same region (Bay of Biscay), keeping the identity at species level would increase the dissimilarity between cells and perhaps provide a more accruate assessment for the ML model. The accuracy of my random forest did show a stark improvement so I continued with the species level investigation. Further to this I decided, that leaving my cells in with only 3 dimensions would also create noise in my model. Its well known that both fish communities and environmental variables stratify with depth. So I've included 'depth bins' in my lat and long cells, where my fish observations will be classified into three depth groups [Shallow, Medium, Deep] the specifics of which you can observe in the code. 

In this script I am to apply both **Unsupervised machine learning**  to my data set in order to idetnfiy and characterise communities from my fish observation data. Once I've idetnified the number of communities using unsupervised machine learning (I'll show both techniques), I will then charactersie the composition of my communities. 

Following this step, I will then apply **Supervised machine learning** in the form of a Random Forest Model. This I will train on the data and then test in order to identify how well the model predicts the community present from the environmental conditions. I will then encode some hyperparamter tuning to maximise my models performance and then finally produce Partial Dependance Plots (PDPs), to help visualise the marginal influence of each of my continuous predictor variables. 

#### Step 1. Load in data and prepare for ML

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import re
from threadpoolctl import threadpool_limits
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GroupKFold, cross_val_predict
from sklearn.metrics import classification_report, f1_score, accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from scipy.stats import randint
from joblib import Memory
import os, tempfile
from sklearn.model_selection import cross_val_predict
from sklearn.inspection import PartialDependenceDisplay
import os
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# Load up the cleaned focus DF I saved before, I'll call it obs_data to start. 
# Identify what region to start/train the algorithm. and create depth bins.
obs_data = pd.read_csv("data/data_interim/cleaned_focus_dataset.csv")
start_region = "fr_atlantic"
depth_bins = [0, 30, 80, 200]
depth_labels = ["Shallow", "medium" , "deep"]

# What columns do I want to work with 
key_cols = [
    "lat","lon","month_year","eventdate", "family","ind_count","fish_depth","region", 
    "surveyID","sst_actual","sst_anomaly","chl","coast_km","ocean_depth"
]

# Run a quick loop to check I've got all my columns 
missing_cols = [c for c in key_cols if c not in obs_data.columns]
if missing_cols:
    raise KeyError(f"Missing required columns in obs_data: {missing_cols}")

#print(obs_data.head(3))

# Set the date columns to correct format then derive specific month and years into seperate columns
obs_data["month_year"] = pd.to_datetime(obs_data["month_year"], errors = "coerce")
obs_data["eventdate"] = pd.to_datetime(obs_data["eventdate"], errors = "coerce")

if "month_year" in obs_data.columns and obs_data["month_year"].notna().any():
    obs_data["year"]  = obs_data["month_year"].dt.year
    obs_data["month"] = obs_data["month_year"].dt.month
elif "eventdate" in df.columns and df["eventdate"].notna().any():
    obs_data["year"]  = obs_data["eventdate"].dt.year
    obs_data["month"] = obs_data["eventdate"].dt.month
else:
    raise ValueError("Need either date column available to derive year/month.")

# Format all numeric columns
numeric_cols = [
    "lat","lon","ind_count","fish_depth","sst_actual","sst_anomaly",
    "chl","coast_km","ocean_depth"
]

for c in numeric_cols:
    if c in obs_data.columns:
        obs_data[c] = pd.to_numeric(obs_data[c], errors="coerce")

# All rows have an individual count as 1 unless otherwise specified
obs_data["ind_count"] = obs_data["ind_count"].fillna(1).astype(int)
# Ind counts are integers
obs_data["ind_count"] = obs_data["ind_count"].astype(int)

# Generate my depth bins
obs_data = obs_data.dropna(subset=["fish_depth"]).copy()
obs_data = obs_data[(obs_data["fish_depth"] >= depth_bins[0]) & (obs_data["fish_depth"] < depth_bins[-1])]
obs_data["depth_bin"] = pd.cut(obs_data["fish_depth"], bins=depth_bins, labels=depth_labels, right=False)

# I'll now just repeat the process I did in script 4. 
grid_size = 0.05

def assign_cells(obs_data, lat='lat', lon='lon', res_deg=0.05):
    x = obs_data.copy()
    x['lat_bin'] = np.floor(x[lat] / res_deg) * res_deg
    x['lon_bin'] = np.floor(x[lon] / res_deg) * res_deg
    x['cell_id'] = x['lat_bin'].round(4).astype(str) + "_" + x['lon_bin'].round(4).astype(str)
    return x

# use my function to create the gridded cell data frame
gridded_cell_df = assign_cells(obs_data, res_deg = grid_size)

# Now I've got my df set up in the cell format, I create a unique identifier for each cell.

gridded_cell_df["rep_ID"] = (
    gridded_cell_df["cell_id"].astype(str)
    + "|" + gridded_cell_df["year"].astype(str)
    + "-" + gridded_cell_df["month"].astype(str).str.zfill(2)
    + "|" + gridded_cell_df["depth_bin"].astype(str)
)

# I'm focusing my machine learning on the french atlantic region due to its high sampling density 
gridded_cell_df = gridded_cell_df[gridded_cell_df["region"] == start_region].copy()
```

```python
# Split up my species name column
gr_cl_sp_df = (
    gridded_cell_df["sp.name"]
    .astype(str)
    .str.strip()
    .str.replace(r'\s+', ' ', regex=True)
)

# I'll build a function to help me split my species name strings into tokens (words), this will remove any extra words added onto the sp name
def first_two_tokens(s):
    toks = s.split()
    if len(toks) >= 2:
        return f"{toks[0].capitalize()} {toks[1].lower()}"
    return np.nan

gridded_cell_df['sp'] = gr_cl_sp_df.apply(first_two_tokens)
gridded_cell_df = gridded_cell_df.dropna(subset=['sp'])

# Total up my species found in each cell. 
rep_species = (
    gridded_cell_df.groupby(['rep_ID','sp'], as_index=False)['ind_count']
      .sum()
      .rename(columns={'ind_count':'abund'})
)

# Pivot my table to make species into columns
comm_counts = (
    rep_species
    .pivot_table(index='rep_ID', columns='sp', values='abund',
                 aggfunc='sum', fill_value=0)
    .sort_index()
)
# Quick check
comm_counts.head(10)

# Filter out my rare species 
min_cells = 20
occ = (comm_counts > 0).sum(axis=0)
comm_counts = comm_counts.loc[:, occ >= min_cells]

# Apply a Hellinger transformation in order to account for any large species count numbers from skeweing model. 
row_sums = comm_counts.sum(axis=1).replace({0: np.nan})
hellinger = np.sqrt(comm_counts.div(row_sums, axis=0)).fillna(0)

print("hellinger transformed matrix shape:", hellinger.shape)
display(hellinger.head(10))

# Now I'm just going to keep my metadata going forward in a seperate site_meta df for later
site_meta = (
    gridded_cell_df
    .drop_duplicates(subset=["rep_ID"])
    .loc[:, ["rep_ID", "cell_id", "lat", "lon", "year", "month", "depth_bin", "region" , "sst_actual", "sst_anomaly", "chl", "coast_km", "ocean_depth"]]
    .copy()
)

site_meta = site_meta.set_index("rep_ID")
```

#### Step 2. Unsupervised Machine Learning in order to identify 'Communities'

How many Clusters ???? 
There are two strong methods for this age old question. The first is two determine how the within cluster variance drops as the number of clusters increases - Known as the Elbow Method (no idea why elbow). The second is two quantify how well each of my replicates, sits within its identified cluster - Known as the silhouette score. For this next section we'll need a few specialised packages from sklearn.

```python
# define the range for number of clusters (k), lets say my max numbre of clusters is going to be 15
K_range = range(2,15)

# make vectors for youre inter-cluster variance (icv) and silhouette scores (sil_scores)
icvs = []
sil_scores = []

# Create a loop to calculate values for icv and sil_scores across the range of clusters groups i've specified. random state = arthur dent, lets try 20 init
# the threadpool inclusion at the top is simply to stop windows giving me this warning that keeps popping up
with threadpool_limits(limits=1):
    for k in K_range:
        model = KMeans(n_clusters=k, random_state=42, n_init=20)
        model.fit(hellinger)
        icvs.append(model.inertia_)
        sil_scores.append(silhouette_score(hellinger, model.labels_))

# Generate my elbow method plot
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(K_range, icvs, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-cluster variance')

# Generate my silhouette plot
plt.subplot(1,2,2)
plt.plot(K_range, sil_scores, marker='o', color='green')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Average silhouette score')

plt.tight_layout()
plt.show()
plt.savefig("figures/Elbows_and_silhouettes.jpeg", dpi=300, bbox_inches="tight")
```

The Elbow method is showing that as you increase K the within cluster variance is decreasing. Well of course it would. The trick is to determine at what point does the 'curve' appear. After the curve you can observe the law of diminishing returns as increasing the number of clusters no longer drastically reduces the inter cluster variance. From the above plot it seems between k= 4 and 6 the angle of the curve really picks up. Suggesting that the elbow method as returned between 4 and 6 clusters.

The Silhouette score quanitifes how well each replicate fits in its cluster. 1 being very well, 0 in between clusters, - numbers means somethings wrong. With fish communitiies perfect clustering is a myth, at least in my experience. Usually a score of between 0.3 above is acceptable. Again I'm looking for the number of clusters where the angle of the curve really changes which seems to be around 5.

#### Step 3. Use KMeans with my observed number of clusters to determine community labels 

```python
# Fit KMeans onto my hellinger matrix
k = 5 
kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
labels = kmeans.fit_predict(hellinger)
label_names = [f"C{i+1}" for i in labels]

# Now I'll attach my labels to my replicates
clusters = (
    pd.DataFrame({
        "rep_ID" : hellinger.index, 
        "community_type" : label_names, 
        "community_int" : labels
    })
    .set_index("rep_ID")
)

# I'll look how the reps fall into clusters
print("cluster counts:")
print(clusters["community_type"].value_counts().sort_index().to_string())

# I want to see which families dominate each cluster. use lambda
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=hellinger.columns, index=[f"C{i+1}" for i in range(k)])
top_families = (
    centroids
      .apply(lambda s: s.sort_values(ascending=False).head(5).index.tolist(), axis=1)
      .rename("top_5_species")
      .to_frame()
)

print("Top 5 species for each of my clusters:")
display(top_families)

# Now I'll bring back my site_meta I've saved earlier and attach the labels for mapping later
rep_labels = (
    site_meta
    .join(clusters, how="left")
    .reset_index()
)

print(f"Labeled replicates: {len(rep_labels)}")
display(rep_labels.head())
```

Now that I have quantified my community type and defined theyre composition (sort of). I'm now going to summarise the environmental conditions associated with each community so I can interpret their ecology.

#### Step 5. Supervsied Machine Learning Model

Now that I've used unsupervised learning to determine the number of community types in this dataset, I want to train a supervised learning model to predict fish community types based on environmental, spatial and temporal conditions. I also will aim to understand which environmental factors best predict community compositon.

```python
# load in my dataset
ml_ready_df = pd.read_csv("data/data_interim/dataset_for_supervised_ML.csv")

# Define what are my predictors and what are my target variables
target = "community_type"
predictors = [
    "sst_actual", "sst_anomaly", "chl", "coast_km", "ocean_depth",
    "lat", "lon", "year", "month", "depth_bin", "region"
]

x = ml_ready_df[predictors].copy()
y = ml_ready_df[target].copy()

# encode categorical variables, as my ML model cannot handle the text component. get dummies by pandas is a good way to handle this
x = pd.get_dummies(x, columns=["region","depth_bin", "month", "year"], drop_first=True)

# Now we do our train , test , split 80 - 20. the key part is keeping stratify to y (target) keeps proportion when splitting
x_train , x_test, y_train , y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)


# Fit my randon forest. 
random_forest = RandomForestClassifier(
    n_estimators=400, 
    max_depth=None, 
    min_samples_leaf=2, 
    n_jobs=-1, 
    random_state=42
)
random_forest.fit(x_train, y_train)

# assign my predictions to a vector
y_predict = random_forest.predict(x_test)
accuracy = accuracy_score(y_test, y_predict)

print(f"Accuracy:{accuracy:.3f}\n")
      
print("Classification report (per community):")
print(classification_report(y_test, y_predict, digits=3))

# Im also going to make a confusion matrix to go with the outcome
cm = confusion_matrix(y_test, y_predict, labels=sorted(y.unique()))
cm_df = pd.DataFrame(cm, index=[f"true_{c}" for c in sorted(y.unique())],
                        columns=[f"pred_{c}" for c in sorted(y.unique())])
display(cm_df)

# Feature importance (sorted top 15)
featured = pd.Series(random_forest.feature_importances_, index=x_train.columns).sort_values(ascending=False)
print("\nTop feature importances:")
display(featured.head(15).round(4))
```

Above are my Random forests output results. 
It states an overall model accuracy of 0.453. Originally I had the fish occurences on the family level but it gave a low accuracy of ~0.3 so this is definitely an improvement. However by looking at the F1 scores (Macro 0.321, Weighted = 0.387) it is clear that the model performance varies between classes. For my C1 you can see a recall of 0.83 meaning the model is very good at finding C1, however its precision states around 50% of its predictions are wrong. Interestingly at the other end of the spectrum the model never predicts C4. Observe the confusion matrix, it shows that true C4 are mostly predicted as C1. From the patterns in the confusion matrix, it seems that C1 seems to attract reps from other communities (mostly C4 and C5), C2 splits C1 and C3 which implys it sits between those in feature space. C3 seems to be the most distinct from the C1 attractor as it retains the most of its true reps. 
Again SST, CHl, distance to coast and bathymetry show greatest importance. The month and depth bins much less important. 

From these results I would suggest that the model is biased toward predicting C1 which suggests an overlap or imbalance of my community definitions. Furthemore, the model is not able to predict C4 with current features. 

I'm going to make a few tweaks. First I'll try creating spatial blocks, which will be groups of my lat lon cells in close proximity so that the model can memorize the spatial patterns and better estimate. I'll also assign my months to a cyclic pattern. Once I've made these tweaks, I'll then generate a baseline model in order to define how accurate my model is prior to any tuning. 

Following creation of my baseline model I will then undertake some hyperparameter tuning. 

```python
####### Create my Spatial Blocks ###########################################################

# Make a copy of my data for calibrating before baseline model
df_cal = ml_ready_df.copy() 

# Cyclic month 
df_cal['month_sin'] = np.sin(2*np.pi*df_cal['month']/12)
df_cal['month_cos'] = np.cos(2*np.pi*df_cal['month']/12)

# Spatial blocks ~0.5Â°
df_cal['lat_blk'] = (df_cal['lat'] // 0.5).astype(int)
df_cal['lon_blk'] = (df_cal['lon'] // 0.5).astype(int)
df_cal['spatial_block'] = df_cal['lat_blk'].astype(str) + "_" + df_cal['lon_blk'].astype(str)

# Now define my target variable, numeric features and categorical features
target = 'community_type'
num_feats = ['sst_actual','sst_anomaly','chl','coast_km','ocean_depth',
             'lat','lon','month_sin','month_cos']
cat_feats = ['depth_bin'] 

X_num = df_cal[num_feats]
X_cat = pd.get_dummies(df_cal[cat_feats], drop_first=True) if cat_feats else None
X = pd.concat([X_num, X_cat], axis=1) if X_cat is not None else X_num
y = df_cal[target]
groups = df_cal['spatial_block']

############################################################################################
###### Create spatial generalization basline ###############################################

# Define what are my features and groups
X = X  
y = df_cal['community_type']
groups = df_cal['spatial_block']

numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = [c for c in X.columns if c not in numeric_cols]

# Preprocess my columns, specify not to touch my numeric columns but to convert categories into a dummy varaibles (binary).
# Basically converts my categorical columns into a form that my model can understand and work with
preprocess = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numeric_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_cols)
    ],
    remainder="drop"
)

# Generate my baseline random forest model
baseline_rf = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", RandomForestClassifier(
        n_estimators=400,
        max_depth=None,
        min_samples_leaf=1,
        max_features="sqrt",
        class_weight="balanced",
        n_jobs=-1,
        random_state=42
    ))
])

gkf = GroupKFold(n_splits=5)

# Group-aware out-of-fold predictions
y_pred_cv = cross_val_predict(
    baseline_rf, X, y, cv=gkf.split(X, y, groups), n_jobs=-1, method="predict"
)

print("Baseline (GroupKFold, class-balanced RF)")
print("  Accuracy  :", round(accuracy_score(y, y_pred_cv),3))
print("  Macro F1  :", round(f1_score(y, y_pred_cv, average="macro"),3))
print("  Weighted F1:", round(f1_score(y, y_pred_cv, average="weighted"),3))
print("\nPer-class report:\n", classification_report(y, y_pred_cv, digits=3))

#### basline macro-F1 = 0.317. Thats my baseline 
```

```python
########################################################################################
############## Hyper parameter tuning ##################################################

# reuse my columns specified before
numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = [c for c in X.columns if c not in numeric_cols]

# Cache to speed repeated preprocessing across folds
cache_dir = os.path.join(tempfile.gettempdir(), "sk_cache_biscay")
memory = Memory(cache_dir, verbose=0)

# Again parse my columns to in essence convert all into a numeric format
preprocess = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numeric_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_cols)
    ],
    remainder="drop"
)

# Build my random forest model, ensure it automatically adjusts the weighing of each community based on their frequency
rf = RandomForestClassifier(
    class_weight="balanced",
    n_jobs=-1,
    random_state=42
)

# Now I want to combine all the preprocessing and modelling into one unified workflow/pipeline. this way when I call fit() the preprocess will 
# automatically run and then feed my transformed data into my random forest models. 
pipe = Pipeline(
    steps=[("prep", preprocess), ("clf", rf)],
    memory=memory
)

# Lightweight search space. using 'clf__' I specify where sklearn should look inside the pipeline for my hyperparameteres.
param_dist = {
    "clf__n_estimators": randint(250, 900),       
    "clf__max_depth": [None, 12, 20, 30],
    "clf__min_samples_leaf": [1, 2, 4, 6],
    "clf__max_features": ["sqrt", None],
}

# Implement a spatial block cross validation
gkf = GroupKFold(n_splits=5)

# Heres the guts, implement tuning using random combinations of my hyperparameters specifies above
search = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_dist,
    n_iter=16,               
    scoring="f1_macro",
    cv=gkf,
    n_jobs=-1,
    refit=True,
    random_state=42,
    verbose=1
)

# Now I fit my hyperparamter search 
search.fit(X, y, groups=groups)

# Display results
print("Best params:", search.best_params_)
print("Best CV macro-F1:", round(search.best_score_, 3))

# Store my optimum random forest
best_rf = search.best_estimator_
```

```python
########## Honest OOF evalutation with tuned parameters ####################################################

# Rebuild my above pipeline but this time use my tuned and fixed parameters found from the above search
preprocess = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numeric_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_cols)
    ],
    remainder="drop"
)

best_pipe = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", RandomForestClassifier(
        class_weight="balanced",
        n_estimators=523,
        max_depth=None,
        min_samples_leaf=4,
        max_features="sqrt",
        n_jobs=-1,
        random_state=42
    ))
])

# Define my spatial split
gkf = GroupKFold(n_splits=5)

# Generate my 'Out Of Fold' predictions. This trains up 5 models on training blocks and then predicts only for the block not included. 
y_pred_oof = cross_val_predict(
    best_pipe, X, y,
    cv=gkf.split(X, y, groups),
    n_jobs=-1,
    method="predict"
)
# This will give me a single prediciton per row where that row was never in the models training data 'Out Of Fold'

print("Tuned model (GroupKFold OOF):")
print("  Accuracy   :", round(accuracy_score(y, y_pred_oof), 3))
print("  Macro F1   :", round(f1_score(y, y_pred_oof, average='macro'), 3))
print("  Weighted F1:", round(f1_score(y, y_pred_oof, average='weighted'), 3))
print("\nPer-class report:\n", classification_report(y, y_pred_oof, digits=3))
print("Confusion matrix (rows=true, cols=pred):\n", confusion_matrix(y, y_pred_oof))
```

#### Interpretation of tuned parameter model perfomance. 

My overall accuracy dropped, however the macro F1 improved. This actually reflects a better balance across classes as C4 and C5 predictions improved. C1 and C3 so greater seperation with low corss over from predictions from true values between the two. The confusion matrix for C4 and C5 indicates that the two show some ecological or possibly sampling overlap. Some C2 replicates are still being absorbed by C1, which might suggest a low signal strength for C2. 

Next I'll understand why the model is making these distinction and aim to identify which environmnetal gradients seperate variability. This is done by working out the permuatational feature importance. 

```python
#### Permutation Feature Importance
from sklearn.inspection import permutation_importance

# Reuse X, y, groups from earlier cells
numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = [c for c in X.columns if c not in numeric_cols]

# Run my preprocessor again
preprocess = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numeric_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_cols)
    ],
    remainder="drop"
)

# build my RF with the tuned parameters identified in the random search
rf = RandomForestClassifier(
    class_weight="balanced",
    n_estimators=523,
    max_depth=None,
    min_samples_leaf=4,
    max_features="sqrt",
    n_jobs=-1,
    random_state=42
)

pipe = Pipeline([("prep", preprocess), ("clf", rf)])

# build a function to get output feature names from the preprocessor. This will help me in the forthcoming step. 
# Basically the matrix I inputed into the radom forest has more columns than my dataframe so I need to harmonise this
def get_output_names(preprocessor):
    names = []
    for name, trans, cols in preprocessor.transformers_:
        if name == "remainder" and trans == "drop":
            continue
        if trans == "passthrough":
            # cols can be list of names
            names.extend(cols)
        else:
            try:
                # OneHotEncoder case
                fn = trans.get_feature_names_out(cols)
                names.extend(fn)
            except Exception:
                # Fallback: just append cols
                names.extend(cols)
    return list(names)

gkf = GroupKFold(n_splits=5)

##################### Create a spatial permutation importance loop.  ##############################################################################

# prep containers
fold_imps = []

# Build a loop to compute my spatial splits, whilst keeping spatially seperation between test and train rows 
for fold_idx, (tr, te) in enumerate(gkf.split(X, y, groups)):
    X_tr, X_te = X.iloc[tr], X.iloc[te]
    y_tr, y_te = y.iloc[tr], y.iloc[te]

    # Fit full pipeline on training blocks
    pipe.fit(X_tr, y_tr)

    # Get transformed data for my held out block and the find the fitted classifier
    Xt_te = pipe.named_steps["prep"].transform(X_te)       
    clf    = pipe.named_steps["clf"]

    # Names for transformed columns 
    names_fold = get_output_names(pipe.named_steps["prep"])

    # Permutation importance on the classifier using transformed features
    pi = permutation_importance(
        clf, Xt_te, y_te,
        scoring="f1_macro",
        n_repeats=10,
        n_jobs=-1,
        random_state=42
    )

    # Build per-fold importance table 
    fold_df = pd.DataFrame({
        "feature": names_fold,
        "imp_mean": pi.importances_mean,
        "imp_std":  pi.importances_std,
    })
    fold_df["fold"] = fold_idx

    # Collapse binary dummies back to parent categorial feature
    def base_name(col):
        for cat in categorical_cols:
            if col.startswith(cat + "_"):
                return cat
        return col
    fold_df["base_feature"] = fold_df["feature"].map(base_name)

    # Sum one-hot parts within fold
    fold_df = (fold_df.groupby(["base_feature", "fold"], as_index=False)
                      .agg(imp_mean=("imp_mean", "sum"),
                           imp_std=("imp_std", "mean")))
    fold_imps.append(fold_df)

# Now I want to find the fold averaged spatial permuation importance for each variable
imp_all = pd.concat(fold_imps, ignore_index=True)
imp_summary = (imp_all.groupby("base_feature")
                      .agg(mean_imp=("imp_mean", "mean"),
                           sd_across_folds=("imp_mean", "std"),
                           mean_within_fold_sd=("imp_std", "mean"))
                      .sort_values("mean_imp", ascending=False))

print(imp_summary.head(12).round(4))
```

I'll now build some partial dependance plots, to help visulise the marginal effect of each of my top continuous predictors. 

```python
# Ensure base_feature is a column (not index)
imp_plot = imp_summary.reset_index() if "base_feature" not in imp_summary.columns else imp_summary.copy()

# Sort by importance and take top 5
TOP_N = 5
imp_plot = imp_plot.sort_values("mean_imp", ascending=False)
top_imp = imp_plot.head(TOP_N).iloc[::-1] 

# Plot with error bars across folds
fig, axis = plt.subplots(figsize=(7, 5))
axis.barh(
    top_imp["base_feature"],
    top_imp["mean_imp"],
    xerr=top_imp.get("sd_across_folds", None),
    capsize=3
)
axis.set_xlabel("Permutation importance")
axis.set_ylabel("Predictor")
axis.set_title("Top environmental predictors of fish community type")
plt.tight_layout()
plt.savefig("figures/importance_of_env_predictors.jpeg", dpi=300, bbox_inches="tight")
```
